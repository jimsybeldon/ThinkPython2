While familiar with this factorial recursing function,  it still is a head bender.

```python
def factorial(n):
    """Computes factorial of n recursively."""
    if n == 0:
        return 1
    else:
        recurse = factorial(n-1)
        result = n * recurse
        return result
```

    Algorithm due to Srinivasa Ramanujan, from

```python
def estimate_pi():
    """Computes an estimate of pi.

    Algorithm due to Srinivasa Ramanujan, from 
    http://en.wikipedia.org/wiki/Pi
    """
    total = 0
    k = 0
    factor = 2 * math.sqrt(2) / 9801
    while True:
        num = factorial(4*k) * (1103 + 26390*k)
        den = factorial(k)**4 * 396**(4*k)
        
        total += num / den
        term = factor * num/den
        print("iteration ",k)
        if abs(term) < 1e-15:
            break
        k += 1
    
    
    return 1 / (factor * total)

print(estimate_pi())
```
